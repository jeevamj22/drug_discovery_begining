```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
#target search for corona virus
target=new_client.target
target_query=target.search('coronavirus')
targets=pd.DataFrame.from_dict(target_query)
targets
```

```python
#select and retrieve bioactivity data for SARS coronavirus 3C-like protenaise
selected_target=targets.target_chembl_id[4]
selected_target
```

```python
activity=new_client.activity
res=activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
df=pd.DataFrame.from_dict(res)
```

```python
#handling missing data
df2=df[df.standard_value.notna()]
df2
```

```python
#Data pre-processng of the bioactivity data
#labelling compounds as either being active,inactive or intermediate
bioactivity_class=[]
for i in df2.standard_value:
  if float(i)>=10000:
    bioactivity_class.append("inactive")
  elif float(i)<=1000:
    bioactivity_class.append("active")
  else:
    bioactivity_class.append("intermediate")
```

```python
#iterate the molecule_chembl_id to a list
mol_cid=[]
for i in df2.molecule_chembl_id:
  mol_cid.append(i)
```

```python
canonical_smiles=[]
for i in df2.canonical_smiles:
  canonical_smiles.append(i)
```

```python
#iterate standard value to a list
standard_value=[]
for i in df2.standard_value:
  standard_value.append(i)
```

```python
#combine the four lsit into a dataframe
data_tuples=list(zip(mol_cid,canonical_smiles,bioactivity_class,standard_value))
df3=pd.DataFrame(data_tuples,columns=["molecule_chembl_id","canonical_smiles","bioactivity_class","standard_value"])
```
